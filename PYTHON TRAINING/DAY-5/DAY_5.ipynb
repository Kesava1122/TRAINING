Line Chart: Trends of major search engines

```python
import matplotlib.pyplot as plt

# Convert 'Date' to datetime for better handling
data['Date'] = pd.to_datetime(data['Date'], format='%Y-%m')

# 1. Line Chart: Trends of major search engines
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Google'], label='Google', marker='o')
plt.plot(data['Date'], data['Yahoo!'], label='Yahoo!', marker='s')
plt.plot(data['Date'], data['bing'], label='Bing', marker='^')
plt.title('Search Engine Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Market Share (%)')
plt.legend()
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Bar Chart: Market share comparison at the last recorded date

```python
import matplotlib.pyplot as plt

# Convert 'Date' to datetime for better handling
data['Date'] = pd.to_datetime(data['Date'], format='%Y-%m')

# Bar Chart: Market share comparison at the last recorded date
last_date = data.iloc[-1]
search_engines = ['Google', 'Yahoo!', 'bing', 'Baidu', 'DuckDuckGo']
shares = [last_date[se] for se in search_engines]

plt.figure(figsize=(8, 6))
plt.bar(search_engines, shares, color='skyblue')
plt.title(f'Market Share Comparison ({last_date["Date"].strftime("%Y-%m")})')
plt.ylabel('Market Share (%)')
plt.show()
```

Pie Chart: Distribution for a specific date

```python
import matplotlib.pyplot as plt

# Convert 'Date' to datetime for better handling
data['Date'] = pd.to_datetime(data['Date'], format='%Y-%m')

#  Pie Chart: Distribution for a specific date
specific_date = '2021-01'
filtered_data = data[data['Date'] == specific_date].iloc[0]
pie_values = [filtered_data[col] for col in search_engines]
plt.figure(figsize=(8, 8))
plt.pie(pie_values, labels=search_engines, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab10.colors)
plt.title(f'Search Engine Market Share Distribution ({specific_date})')
plt.show()
```

Histogram: Google share frequency

```python
import matplotlib.pyplot as plt

# Convert 'Date' to datetime for better handling
data['Date'] = pd.to_datetime(data['Date'], format='%Y-%m')

# Histogram: Google share frequency
plt.figure(figsize=(8, 6))
plt.hist(data['Google'], bins=10, color='green', edgecolor='black', alpha=0.7)
plt.title('Distribution of Google Market Share')
plt.xlabel('Market Share (%)')
plt.ylabel('Frequency')
plt.show()
```

 Box Plot: Distribution of market shares

```python
import matplotlib.pyplot as plt

# Convert 'Date' to datetime for better handling
data['Date'] = pd.to_datetime(data['Date'], format='%Y-%m')
# Box Plot: Distribution of market shares
plt.figure(figsize=(12, 6))
data_to_plot = [data[se] for se in search_engines]
plt.boxplot(data_to_plot, labels=search_engines)
plt.title('Market Share Distribution Across Search Engines')
plt.ylabel('Market Share (%)')
plt.show()
```

Matplotlib - Graphs

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load and preprocess data
data = pd.read_csv('/content/search_engine_data.csv')
data['Date'] = pd.to_datetime(data['Date'], format='%Y-%m')  # Convert Date column to datetime

# 1. Matplotlib - Graphs
plt.plot(data['Date'], data['Google'], label='Google'); plt.title('Google Market Share Trend'); plt.legend(); plt.show()  # Line Graph
```

BAR PLOT


```python
plt.bar(data['Date'][:10], data['Yahoo!'][:10]); plt.title('Yahoo! Market Share (First 10 Months)'); plt.show()  # Bar Graph
```

HISTOGRAM

```python
plt.hist(data['bing'], bins=10, color='orange'); plt.title('Bing Market Share Distribution'); plt.show()  # Histogram

```

SEABORN VISUALIZATION LINE PLOT

```python
#  Seaborn Visualizations
sns.lineplot(x=data['Date'], y=data['Google']); plt.title('Google Trend (Seaborn)'); plt.show()  # Line Plot

```

BOX PLOT

```python
sns.boxplot(data=data[['Google', 'Yahoo!', 'bing']]); plt.title('Box Plot of Major Search Engines'); plt.show()  # Box Plot
```

Binomial Distribution

```python
# 3. Binomial Distribution
binom_data = np.random.binomial(n=10, p=0.5, size=1000); sns.histplot(binom_data, bins=10, kde=False, color='green'); plt.title('Binomial Distribution'); plt.show()  # Binomial Histogram

```

Correlation Matrix

```python
# 4. Correlation Matrix
sns.heatmap(data.corr(), annot=True, cmap='coolwarm'); plt.title('Correlation Matrix'); plt.show()  # Heatmap for Correlation Matrix
```
